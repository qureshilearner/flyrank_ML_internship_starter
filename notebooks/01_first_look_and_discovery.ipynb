## 2. Discover a real truth yourself

The safest, most satisfying early wins are **things you find in the data** — un-leakable, and they *are* the core lesson. Run the three cells below. Each is ~10 lines of pandas and each overturns a common SEO belief.

Every number is **computed live from the shipped CSV** — nothing is hardcoded.

```python
import pandas as pd, numpy as np
df = pd.read_csv("data/raw/content_refresh_anonymized.csv")
print(df.shape[0], "rows,", df.shape[1], "columns")
df.head(3)
```

### Discovery A — "High search volume means more traffic." Does it?

```python
corr = df["search_volume"].corr(df["impressions_90d"])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B — the CTR cliff by position
Click-through rate is not flat: it collapses as you move down the results.

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("position_tier")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C — is longer content the lever?
Compare word count for **declining** vs **growing** pages.

```python
wc = df.groupby("trend_direction")["word_count"].median()
print(wc.round(0).to_string())
print("\n'down' vs 'up' pages have almost the same median word count -> length is not the lever.")
```

### Discovery A — "High search volume means more for pages with `impressions_90d > 0` — does the correlation change?

```python

# filtering the columns first
df_impressions = df[df["impressions_90d"] > 0]

corr = df["search_volume"].corr(df_impressions['impressions_90d'])
print(f"Correlation between search_volume and impressions_90d: {corr:.3f}")
print("Near zero -> keyword search volume barely predicts the traffic a page actually gets.")
```

### Discovery B
which content_type has the worst CTR at the same position tier?

```python
visible = df[df["impressions_90d"] >= 100]
ctr_by_pos = visible.groupby("content_type")["ctr"].mean().sort_values(ascending=False)
print(ctr_by_pos.round(4).to_string())
ctr_by_pos.plot(kind="bar", title="Mean CTR by position tier (impressions >= 100)", ylabel="CTR");
```

### Discovery C
Find another belief to test: does content_age_days relate to trend_direction?\
Does avg_position relate to ctr?

```python
import pandas as pd
import matplotlib.pyplot as plt

# content_age_days vs trend_direction
print("=== Content Age vs Trend Direction ===")
print(df.groupby("trend_direction")["content_age_days"].agg(['mean', 'median', 'count']))

# avg_position vs ctr
print("\n=== Average Position vs CTR ===")
corr_position_ctr = df["avg_position"].corr(df["ctr"])
print(f"Correlation: {corr_position_ctr:.4f}")

# Visualize both
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# boxplot for content_age_days by trend_direction
df.boxplot(column="content_age_days", by="trend_direction", ax=axes[0])
axes[0].set_title("Content Age by Trend Direction")

# scatter for avg_position vs ctr
axes[1].scatter(df["avg_position"], df["ctr"], alpha=0.5)
axes[1].set_xlabel("Average Position")
axes[1].set_ylabel("CTR")
axes[1].set_title(f"Position vs CTR (r={corr_position_ctr:.3f})")

plt.tight_layout()
plt.show()
```
